```python
#--------------------------------------------------------Part (a)----------------------------------------------------------------

#Import required libraries
import pandas as pd
import string
import warnings
warnings.filterwarnings('ignore')

#Read and merge two data sets
train = pd.read_csv('hotel_train.csv', header=None)
test = pd.read_csv('hotel_test.csv', header=None)
df_hotel_review = pd.concat([train, test])

#Add column names to the merged data set
df_hotel_review.columns=['Review Number','Review','Opinion']

#No of total and unique words before cleaning
print("--------------Number of words before cleaning---------------")
#No of total words
print("Number of total words: ", df_hotel_review['Review'].str.lower().str.split().str.len().sum())

#No of unique words
results = set()
df_hotel_review['Review'].str.lower().str.split().apply(results.update)
print("Number of unique words: ", len(results))


#Clean data
def clean_text(text):
    # lower text
    text = text.lower()
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # remove words that contain numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    # remove empty tokens and words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
    text = " ".join(text)
    return(text)

# clean text data continued....
df_hotel_review["review_clean"] = df_hotel_review["Review"].apply(lambda x: clean_text(x))
df_hotel_review["review_clean"]

#Add new column as 'Feedback_Length' which provides the length of a string of 'Feedback' column
df_hotel_review['Review_Length']  = df_hotel_review['review_clean'].str.len()
df_hotel_review

#Drop rows where feedback length <= 4
df_hotel_review = df_hotel_review.drop(df_hotel_review[df_hotel_review.Review_Length < 5].index)

#No of posts which are suggestions and not suggestions
print(df_hotel_review.groupby('Opinion').size()) # 1 -> suggestions 0-> not suggestions

import seaborn as sns
# Now lets check the class distributions
sns.countplot("Opinion",data=df_hotel_review)

#Replace unwanted charaters with empty string
df_hotel_review["Clean_Feedback"] = df_hotel_review['review_clean'].astype(str).str.lower().str.replace('[^\w\s]','')

print("--------------Number of words after cleaning---------------")
#No of total words
print("Number of total words: ", df_hotel_review['Clean_Feedback'].str.lower().str.split().str.len().sum())

#No of unique words
results = set()
df_hotel_review['Clean_Feedback'].str.lower().str.split().apply(results.update)
print("Number of unique words: ", len(results))
```

```python
#--------------------------------------------------------- Part (c)-------------------------------------------------------------

#Convert lemmatize list to the flat list
def generate_flat_list(list):
    lemmatize_list = [[' '.join(i)] for i in list]
    flat_lemmatize_list = []
    for sublist in lemmatize_list:
        for item in sublist:
            flat_lemmatize_list.append(item)
    return flat_lemmatize_list

df_hotel_review['lemmatize_strings'] = generate_flat_list(df_hotel_review['lemmatize'])

# Divide the data into training and testing sets
from sklearn.model_selection import train_test_split

def prepare_datasets(corpus, labels, test_data_proportion):
    train_X, test_X, train_Y, test_Y = train_test_split(corpus, labels, 
                                                        test_size=test_data_proportion)
    return train_X, test_X, train_Y, test_Y

#BOW Representation
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()
bag_of_words = count_vectorizer.fit_transform(df_hotel_review['lemmatize_strings'])
bag_of_words = pd.DataFrame(bag_of_words.toarray(),columns = count_vectorizer.get_feature_names())

#Split bag of words into training and testing sets
bog_train_corpus, bog_test_corpus, bog_train_labels, bog_test_labels = prepare_datasets(bag_of_words,
                                                                       df_hotel_review['Opinion'],
                                                                        0.33)

#TDIDF representation
from sklearn.feature_extraction.text import TfidfVectorizer

v = TfidfVectorizer()
x = v.fit_transform(df_hotel_review['lemmatize_strings'])
tfidf = pd.DataFrame(x.toarray(), columns=v.get_feature_names())

#Split TDIDF into training and testing sets
tfidf_train_corpus, tfidf_test_corpus, tfidf_train_labels, tfidf_test_labels = prepare_datasets(tfidf,
                                                                       df_hotel_review['Opinion'],
                                                                 0.33)

# Define function to calculate the 4 common mertics
def get_metrics(true_labels, predicted_labels):
    
    print('Accuracy:', np.round(metrics.accuracy_score(true_labels, predicted_labels),4))
    print('Precision:', np.round(metrics.precision_score(true_labels, predicted_labels,average='weighted'),4))
    print('Recall:', np.round(metrics.recall_score(true_labels, predicted_labels, average='weighted'),4))
    print('F1 Score:', np.round(metrics.f1_score(true_labels, predicted_labels, average='weighted'),4))
                        
# predict the results and evaluate predictions against the test data
def train_predict_evaluate_model(classifier, train_features, train_labels, test_features, test_labels):
    # build model    
    classifier.fit(train_features, train_labels)
    # predict using model
    predictions = classifier.predict(test_features) 
    # evaluate model prediction performance   
    get_metrics(true_labels=test_labels, 
                predicted_labels=predictions)
    return predictions

# Import the two classification algorithms 
# Based on the features we extract, we altogether have 4 combinations of models to train               
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
import numpy as np
from sklearn import metrics
from sklearn import svm
import seaborn as sns
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt

mnb = MultinomialNB(alpha=1)
svm = svm.SVC(C=1, kernel='linear')

def plot_confuison_matrix(y_test, y_pred):
    cf_matrix = confusion_matrix(y_test, y_pred)
    ax = sns.heatmap(cf_matrix/np.sum(cf_matrix), annot=True, fmt='.2%', cmap='Reds')
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ')
    plt.show()

# Multinomial Naive Bayes with bag of words features
print("----- MNB BOW Performance ------\n")
mnb_bow_predictions = train_predict_evaluate_model(classifier=mnb,
                                           train_features=bog_train_corpus,
                                           train_labels=bog_train_labels,
                                           test_features=bog_test_corpus,
                                           test_labels=bog_test_labels)
print("MNB_bow_predictions: ", mnb_bow_predictions)
print("\n---------Confusion matrix for MNB BOW model-------")
plot_confuison_matrix(bog_test_labels, mnb_bow_predictions)

# Support Vector Machine with bag of words features
print("----- SVM BOW Performance ------\n")
svm_bow_predictions = train_predict_evaluate_model(classifier=svm,
                                           train_features=bog_train_corpus,
                                           train_labels=bog_train_labels,
                                           test_features=bog_test_corpus,
                                           test_labels=bog_test_labels)
print("SVM_bow_predictions: ", svm_bow_predictions)
print("\n---------Confusion matrix for SVM BOW model-------")
plot_confuison_matrix(bog_test_labels, svm_bow_predictions)              

# Multinomial Naive Bayes with tfidf features
print("----- MNB tfidf Performance ------\n")
mnb_tfidf_predictions = train_predict_evaluate_model(classifier=mnb,
                                           train_features=tfidf_train_corpus,
                                           train_labels=tfidf_train_labels,
                                           test_features=tfidf_test_corpus,
                                           test_labels=tfidf_test_labels)
print("MNB_tfidf_predictions: ", mnb_tfidf_predictions)
print("\n---------Confusion matrix for MNB tfidf model-------")
plot_confuison_matrix(tfidf_test_labels, mnb_tfidf_predictions)


# Support Vector Machine with tfidf features
print("----- SVM tfidf Performance ------\n")
svm_tfidf_predictions = train_predict_evaluate_model(classifier=svm,
                                           train_features=tfidf_train_corpus,
                                           train_labels=tfidf_train_labels,
                                           test_features=tfidf_test_corpus,
                                           test_labels=tfidf_test_labels)
print("SVM_tfidf_predictions: ", svm_tfidf_predictions)

# Confusion matrix for the tfidf-based SVM model 


#Introduce sub plot here
print("\n---------Confusion matrix for SVM tfidf model-------")
plot_confuison_matrix(tfidf_test_labels, svm_tfidf_predictions)
```

```python
#--------------------------------------------------------- Part (d)-------------------------------------------------------------
#Clean data
def clean_text(text):
    # lower text
    text = text.lower()
    # tokenize text and remove puncutation
    text = [word.strip(string.punctuation) for word in text.split(" ")]
    # remove words that contain numbers
    text = [word for word in text if not any(c.isdigit() for c in word)]
    # remove empty tokens and words with only one letter
    text = [t for t in text if len(t) > 1]
    # join all
    text = " ".join(text)
    return(text)

# clean text data
df_hotel_review["review_clean"] = df_hotel_review["Review"].apply(lambda x: clean_text(x))

#Add new column as 'Feedback_Length' which provides the length of a string of 'Feedback' column
df_hotel_review['Review_Length']  = df_hotel_review['review_clean'].str.len()

#Drop rows where feedback length <= 4
df_hotel_review = df_hotel_review.drop(df_hotel_review[df_hotel_review.Review_Length < 5].index)

#Replace unwanted charaters with empty string
df_hotel_review["Clean_Feedback"] = df_hotel_review['review_clean'].astype(str).str.lower().str.replace('[^\w\s]','')


#Using gensim as an open-source software library that use modern statistical machine learning to remove stop words
#--------Performance improvements step 01------- 
#gensim library has 337 stop words which is more than the stop words in nltk library which is 179.
#Try with gensim library instead of nltk
import gensim
from gensim.parsing.preprocessing import STOPWORDS
import wordnet, nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
from sklearn import svm

def exclude_stop_words_using_gensim(text):
    stop = STOPWORDS
    return text.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

def exclude_stop_words(text):
    stop = stopwords.words('english')
    return text.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

def tokenize(text):
    return text.apply(word_tokenize)

#Lemmatize with pos tags 
#--------Performance improvements step 02------- 
#POS tags are used to describe the lexical terms that we have within our text.
# Annotate text tokens with POS tags
def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)
    
# Lemmatize text based on POS tags using Wordnet
def lemmatize_text(text):
    lmtzr = WordNetLemmatizer()
    return text.apply(lambda lst:[lmtzr.lemmatize(word, get_wordnet_pos(word)) for word in lst])

# (i)
#Exclude stop words
df_hotel_review['Feedbacks_without_stopwords'] = exclude_stop_words_using_gensim(df_hotel_review['Clean_Feedback'])

#Get word count of each row based on 'Feedbacks_without_stopwords' column
df_hotel_review['Feedback_without_stopwords_Length']  = df_hotel_review['Feedbacks_without_stopwords'].str.split().str.len()

#After removing stop words some records have 'Feedback_without_stopwords_Length' has zero. we can drop those records as well
df_hotel_review = df_hotel_review.drop(df_hotel_review[df_hotel_review.Feedback_without_stopwords_Length == 0].index)

#TOKENIZATION
df_hotel_review['Tokenized_Reviews'] = tokenize(df_hotel_review['Feedbacks_without_stopwords'])

#Lemmatize
df_hotel_review['lemmatize'] = lemmatize_text(df_hotel_review['Tokenized_Reviews'])

df_hotel_review['lemmatize_strings'] = generate_flat_list(df_hotel_review['lemmatize'])


#BOW Representation
from sklearn.feature_extraction.text import CountVectorizer

#Including (1,3)-grams and other parameters as features boosted models predictive power for bag of word representation
#--------Performance improvements step 03------- 
count_vectorizer = CountVectorizer(stop_words='english', ngram_range=(1,3), min_df=2)
bag_of_words = count_vectorizer.fit_transform(df_hotel_review['lemmatize_strings'])

#Split bag of words into training and testing sets
bog_train_corpus, bog_test_corpus, bog_train_labels, bog_test_labels = prepare_datasets(bag_of_words,
                                                                       df_hotel_review['Opinion'],
                                                                        test_data_proportion=0.3)

#TDIDF representation
from sklearn.feature_extraction.text import TfidfVectorizer

#Including (1,3)-grams and other parameters as features boosted models predictive power for Tfidf representation
#--------Performance improvements step 05-------
v = TfidfVectorizer(stop_words='english', ngram_range=(1,3), min_df=2)
x = v.fit_transform(df_hotel_review['lemmatize_strings'])

#Split TDIDF into training and testing sets
tfidf_train_corpus, tfidf_test_corpus, tfidf_train_labels, tfidf_test_labels = prepare_datasets(x,
                                                                       df_hotel_review['Opinion'],
                                                                        test_data_proportion=0.3)

from sklearn import svm
#Parameter Tuning for two models
#--------Performance improvements step 07-------
mnb = MultinomialNB(alpha=2) #1
svm = svm.SVC(C=3, kernel='linear')

# Multinomial Naive Bayes with bag of words features
print("----- MNB BOW Performance ------\n")
mnb_bow_predictions = train_predict_evaluate_model(classifier=mnb,
                                           train_features=bog_train_corpus,
                                           train_labels=bog_train_labels,
                                           test_features=bog_test_corpus,
                                           test_labels=bog_test_labels)
print("MNB_bow_predictions: ", mnb_bow_predictions)

# Support Vector Machine with bag of words features
print("\n----- SVM BOW Performance ------\n")
svm_bow_predictions = train_predict_evaluate_model(classifier=svm,
                                           train_features=bog_train_corpus,
                                           train_labels=bog_train_labels,
                                           test_features=bog_test_corpus,
                                           test_labels=bog_test_labels)
print("SVM_bow_predictions: ", svm_bow_predictions)
               

# Multinomial Naive Bayes with tfidf features
print("\n----- MNB tfidf Performance ------\n")
mnb_tfidf_predictions = train_predict_evaluate_model(classifier=mnb,
                                           train_features=tfidf_train_corpus,
                                           train_labels=tfidf_train_labels,
                                           test_features=tfidf_test_corpus,
                                           test_labels=tfidf_test_labels)
print("MNB_tfidf_predictions: ", mnb_tfidf_predictions)

# Support Vector Machine with tfidf features
print("\n----- SVM tfidf Performance ------\n")
svm_tfidf_predictions = train_predict_evaluate_model(classifier=svm,
                                           train_features=tfidf_train_corpus,
                                           train_labels=tfidf_train_labels,
                                           test_features=tfidf_test_corpus,
                                           test_labels=tfidf_test_labels)
print("SVM_tfidf_predictions: ", svm_tfidf_predictions)

print("\n---------Confusion matrix for best perform model-------")
plot_confuison_matrix(tfidf_test_labels, svm_tfidf_predictions)
```

```python
#--------------------------------------------------------- Part (e)-------------------------------------------------------------

# An unsupervised approach to categorize the suggestions 
# K means clustering is used as the clustering algorithm here

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# We improve our feature matrix builder with 3 additional optional parameters
# This allows us to extract not only word features, but also n-gram features
# We can also set the minimum and maximum frequencies to be considered as valid
# NB: All these are simply passed on to sklearn's Vectorizer classes
def build_feature_matrix(documents, feature_type='frequency',
                         ngram_range=(1, 1), min_df=1):

    feature_type = feature_type.lower().strip()  
    
    if feature_type == 'binary':
        vectorizer = CountVectorizer(binary=True, min_df=min_df,
                                      ngram_range=ngram_range)
    elif feature_type == 'frequency':
        vectorizer = CountVectorizer(binary=False, min_df=min_df,
                                      ngram_range=ngram_range)
    elif feature_type == 'tfidf':
        vectorizer = TfidfVectorizer(min_df=min_df, 
                                     ngram_range=ngram_range)
    else:
        raise Exception("Wrong feature type entered. Possible values: 'binary', 'frequency', 'tfidf'")

    feature_matrix = vectorizer.fit_transform(documents).astype(float)
    
    return vectorizer, feature_matrix

#Considering only the suggestions
df_sug = df_hotel_review.loc[df_hotel_review['Opinion'] == 1, 'lemmatize_strings']
df_suggestions = pd.DataFrame({'Suggestions':df_sug.values})

# Extract (tf-idf) features using out utils library
vectorizer, feature_matrix = build_feature_matrix(df_suggestions['Suggestions'],
                                                  feature_type='tfidf',
                                                  min_df=2, 
                                                  ngram_range=(1, 3))
#How many features do we have
print("------Feature Matrix Shape---------")
print(feature_matrix.shape) 
#Convert feature matrix to data frame
df_tdidf = pd.DataFrame(feature_matrix.toarray(), columns=vectorizer.get_feature_names())

# Get the names of the features
feature_names = vectorizer.get_feature_names()

# Show the first twenty features
print("------Show the first twenty features------")
print(feature_names[:20])


#--------------determine the optimal number of clusters for k-means clustering---------------
#identify the sum of squared distances of samples to the nearest cluster centre
#Elbow method is used to determine the optimal number of clusters here
from sklearn.cluster import KMeans

Sum_of_squared_distances = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_tdidf)
    Sum_of_squared_distances.append(kmeanModel.inertia_)
plt.figure(figsize=(16,8))
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('The Elbow Method showing the optimal k')
plt.show()

# Define the k-means clustering function which will call sklearn's KMeans function
# with our parameters
def k_means(feature_matrix, num_clusters):
    km = KMeans(n_clusters=num_clusters,
                max_iter=10000)
    km.fit(feature_matrix)
    clusters = km.labels_
    return km, clusters

# -----------Optimal k can be determined as 5 based on the elbow method results--------------------
# We will assume that we want the all reviews divided into 6 clusters (try other values)
num_clusters = 5
# Call the above function to get clusters and assign the cluster labels to the reviews
km_obj, clusters = k_means(feature_matrix=df_tdidf,
                           num_clusters=num_clusters)
df_suggestions['Cluster'] = clusters


from collections import Counter
# Get the total number of reviews per cluster
c = Counter(clusters)
print(c.items())

# We define a generic function to extract details of the clusters we get (from each of our algorithms)
def get_cluster_data(clustering_obj, review_data, 
                     feature_names, num_clusters,
                     topn_features=10):

    cluster_details = {}  
    # Get cluster centroids
    ordered_centroids = clustering_obj.cluster_centers_.argsort()[:, ::-1]
    # Get key features for each cluster
    # Get reviews belonging to each cluster
    for cluster_num in range(num_clusters):
        cluster_details[cluster_num] = {}
        cluster_details[cluster_num]['cluster_num'] = cluster_num
        key_features = [feature_names[index] 
                        for index 
                        in ordered_centroids[cluster_num, :topn_features]]
        cluster_details[cluster_num]['key_features'] = key_features
        
        reviews = review_data[review_data['Cluster'] == cluster_num]['Suggestions'].values.tolist()
        cluster_details[cluster_num]['review'] = reviews
    
    return cluster_details
        
       
# We also define a simple function to print the data we extracted with the above function
def print_cluster_data(cluster_data):
    # Print cluster details
    for cluster_num, cluster_details in cluster_data.items():
        print('Cluster {} details:'.format(cluster_num))
        print('-'*20)
        print('Key features:', cluster_details['key_features'])  
        
# We want to be able to visualize the clusters that we got
# Since these will mostly be for multidimensional data, we need to reduce their dimensionality
# We use sklearn's multidimensional scaling (MDS) algorithm to do this
import matplotlib.pyplot as plt
from sklearn.manifold import MDS
from sklearn.metrics.pairwise import cosine_similarity
import random
from matplotlib.font_manager import FontProperties

# This is not a regular function - it has no return value, but only a 'side effect' (plotting)
# This is a generic function to plot the output of any clustering algorithm with dynamic # of clusters
def plot_clusters(num_clusters, feature_matrix,
                  cluster_data, review_data,
                  plot_size=(16,8)):
    # Generate random colors for clusters                  
    def generate_random_color():
        color = '#%06x' % random.randint(0, 0xFFFFFF)
        return color
    # Define different markers for clusters    
    markers = ['o', 'v', '^', '<', '>', '8', 's', 'p', '*', 'h', 'H', 'D', 'd']
    # Build the cosine distance matrix
    cosine_distance = 1 - cosine_similarity(feature_matrix) 
    # Reduce the dimensionality (in this case to 2) using MDS
    mds = MDS(n_components=2, dissimilarity="precomputed", 
              random_state=1)
    # Get the coordinates of clusters in the new low-dimensional (2-dim) space
    plot_positions = mds.fit_transform(cosine_distance)  
    x_pos, y_pos = plot_positions[:, 0], plot_positions[:, 1]
    # Set up the cluster plotting data
    cluster_color_map = {}
    cluster_name_map = {}
    for cluster_num, cluster_details in cluster_data.items():
        # Assign cluster features to unique label
        cluster_color_map[cluster_num] = generate_random_color()
        cluster_name_map[cluster_num] = ', '.join(cluster_details['key_features'][:5]).strip()
    # Map each unique cluster label with its coordinates and reviews
    cluster_plot_frame = pd.DataFrame({'x': x_pos,
                                       'y': y_pos,
                                       'label': review_data['Cluster'].values.tolist(),
                                       'title': review_data['Suggestions'].values.tolist()
                                        })
    grouped_plot_frame = cluster_plot_frame.groupby('label')
    # Set the figure size and axes of the plot
    fig, ax = plt.subplots(figsize=plot_size) 
    ax.margins(0.05)
    # Plot each cluster with its co-ordinates and the corresponding review
    for cluster_num, cluster_frame in grouped_plot_frame:
         marker = markers[cluster_num] if cluster_num < len(markers) \
                  else np.random.choice(markers, size=1)[0]
         ax.plot(cluster_frame['x'], cluster_frame['y'], 
                 marker=marker, linestyle='', ms=12,
                 label=cluster_name_map[cluster_num], 
                 color=cluster_color_map[cluster_num], mec='none')
         ax.set_aspect('auto')
         ax.tick_params(axis= 'x', which='both', bottom='off', top='off',        
                        labelbottom='off')
         ax.tick_params(axis= 'y', which='both', left='off', top='off',         
                        labelleft='off')
    fontP = FontProperties()
    fontP.set_size('small')    
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.01), fancybox=True, 
              shadow=True, ncol=5, numpoints=1, prop=fontP)        
    plt.show() 

# Analyse the k-means clustering data
cluster_data =  get_cluster_data(clustering_obj=km_obj,
                                 review_data=df_suggestions,
                                 feature_names=df_tdidf.columns.tolist(),
                                 num_clusters=num_clusters,
                                 topn_features=5)         

# Print the detailed clustering analysis results
print_cluster_data(cluster_data)    
    
    
# Now visualize the clusters produced
plot_clusters(num_clusters=num_clusters, 
             feature_matrix=df_tdidf,
             cluster_data=cluster_data, 
             review_data=df_suggestions,
            plot_size=(16,8))  
```

K means custering is used to cluster the data set here. Elbow method is used to select the best optimal clusters here. Each clusters and data belongs to each cluster are shown and graphically represented as above results. Optimal number of clusters are selected as 5 based on the elbow method output. (Elbow method was executed multiple times(around 10 times) and got the most frequent no of clusters as optimal value as 5 here.)
