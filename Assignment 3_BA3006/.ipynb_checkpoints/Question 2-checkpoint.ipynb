```python
#--------------------------------------------------------01-----------------------------------------------------------------

import pandas as pd

dfq1 = pd.read_csv("question_01.csv")
dfq1 = dfq1[['[run number]', 'num-people', 'ticks']]
dfq1.columns = ['Run Number', 'Num People', 'Ticks']
print(dfq1.head())


#--------------------------------------------------------part a--------------------------------------------
meanq1 = dfq1.groupby('Num People').mean()
stdq1 = dfq1.groupby('Num People').std()
mean_sd_df1 = pd.merge(meanq1, stdq1, how='inner', on='Num People').reset_index()
mean_sd_df1.drop(['Run Number_x', 'Run Number_y'], axis=1, inplace=True)
mean_sd_df1.columns = ['Population Density', 'Mean', 'Std']
print(mean_sd_df1)


#--------------------------------------------------------part b (i)--------------------------------------------
# Import libraries
import matplotlib.pyplot as plt

#Define data
x = dfq1['Num People']
y = dfq1['Ticks']
fig = plt.figure(figsize =(10, 5))
 # Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
# Creating plot
plt.scatter(x, y)
plt.title(" Time To Achieve 100% Infection Vs Population")
plt.xlabel('Initial Number of People')
plt.ylabel('Time To Achieve 100% Infection')
plt.show()

#--------------------------------------------------------part b (ii)--------------------------------------------
fig = plt.figure()
x = mean_sd_df1['Population Density']
y = mean_sd_df1['Mean']
error = mean_sd_df1['Std']
fig = plt.figure(figsize =(10, 5))
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
plt.errorbar(x, y, yerr = error, fmt = 'o', ecolor = 'r', color = 'green', marker='o')
plt.title(" Mean Time To Achieve 100% Infection Vs Population")
plt.xlabel('Initial Number of People')
plt.ylabel('Mean Time To Achieve 100% Infection')
plt.show()
```
