```python
##Question (4)

import pandas as pd

df = pd.read_csv("Ass3_Ques4.csv")
df = df[['[run number]', 'density', '[final]']]
df.columns = ['Run Number', 'Relationship Density', 'Ticks']
print(df.head())

mean = df.groupby('Relationship Density').mean()
std = df.groupby('Relationship Density').std()
mean_sd_df = pd.merge(mean, std, how='inner', on='Relationship Density').reset_index()
mean_sd_df.drop(['Run Number_x', 'Run Number_y'], axis=1, inplace=True)
mean_sd_df.columns = ['Relationship Density', 'Mean', 'Std']
print(mean_sd_df)


##Question (2) - part c (i)
# Import libraries
import matplotlib.pyplot as plt

#Define data
x = df['Relationship Density']
y = df['Ticks']
fig = plt.figure(figsize =(10, 5))
 # Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
# Creating plot
plt.scatter(x, y)
plt.title("Tick Count vs Relationship Density")
plt.xlabel('Relationship Density')
plt.ylabel('Tick Count')
plt.show()

##Question (3)part c (ii)
fig = plt.figure()
x = mean_sd_df['Relationship Density']
y = mean_sd_df['Mean']
error = mean_sd_df['Std']
fig = plt.figure(figsize =(10, 5))
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
plt.errorbar(x, y, yerr = error, fmt = 'o', ecolor = 'r', color = 'blue', marker='o')
plt.title("Tick Count vs Relationship Density")
plt.xlabel('Relationship Density')
plt.ylabel('Tick Count')
plt.show()
```
