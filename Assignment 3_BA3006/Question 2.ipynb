```python
##Question (2)

import pandas as pd

df = pd.read_csv("Ass3_Ques2.csv")
df = df[['[run number]', 'social-influence', '[final]']]
df.columns = ['Run Number', 'Social Influence', 'Ticks']
print(df.head())

mean = df.groupby('Social Influence').mean()
std = df.groupby('Social Influence').std()
mean_sd_df = pd.merge(mean, std, how='inner', on='Social Influence').reset_index()
mean_sd_df.drop(['Run Number_x', 'Run Number_y'], axis=1, inplace=True)
mean_sd_df.columns = ['Social Influence', 'Mean', 'Std']
print(mean_sd_df)


##Question (2) - part c (i)
# Import libraries
import matplotlib.pyplot as plt

#Define data
x = df['Social Influence']
y = df['Ticks']
fig = plt.figure(figsize =(10, 5))
 # Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
# Creating plot
plt.scatter(x, y)
plt.title("Tick Count vs Social Influence")
plt.xlabel('Social Influence')
plt.ylabel('Tick Count')
plt.show()

##Question (2)part c (ii)
fig = plt.figure()
x = mean_sd_df['Social Influence']
y = mean_sd_df['Mean']
error = mean_sd_df['Std']
fig = plt.figure(figsize =(10, 5))
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
plt.errorbar(x, y, yerr = error, fmt = 'o', ecolor = 'r', color = 'blue', marker='o')
plt.title("Tick Count vs Social Influence")
plt.xlabel('Social Influence')
plt.ylabel('Tick Count')
plt.show()
```
